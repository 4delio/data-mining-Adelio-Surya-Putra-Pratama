```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.float_format', '{:,.0f}'.format)

url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url)

print("First 5 rows:")
display(df.head())

print("\nDataset Info:")
df.info()

print("\nDescriptive Statistics:")
display(df.describe())
```

# KEY FINDINGS:



*   Harga (price_in_rp) sangat bervariasi, dari ratusan juta hingga puluhan miliar, menunjukkan adanya properti dari kelas menengah hingga premium.

*   Variabel fisik seperti bedrooms, bathrooms, land_size_m2, building_size_m2 bisa digunakan untuk melihat faktor utama yang memengaruhi harga

* Terdapat beberapa kolom dengan missing values cukup banyak misalnya year_built, building_age, building_orientation.



```python
print("\nMissing Values per Column:")
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})
null_df = null_df[null_df['Null Count'] > 0]
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)
null_df
```

```python

num_cols_small = ['bedrooms', 'bathrooms', 'land_size_m2', 'building_size_m2', 'floors']
for col in num_cols_small:
    df[col].fillna(df[col].median(), inplace=True)
cat_cols = ['property_type', 'certificate', 'furnishing', 'property_condition', 'ads_id']
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
df.drop(columns=['building_age', 'year_built', 'building_orientation'], inplace=True)
print(df.isnull().sum())
```

# STRATEGY
Untuk mengatasi Missing values yang ada, saya memutuskan untuk menggunakan modus sebagai metode, dengan menggunakan modus akan menjaga distribusi data agar tetap realistis, karena mengganti nilai kosong dengan nilai paling umum dan tidak ada data yang dibuang, sehingga dataset tetap utuh.

```python
num_cols = ["price_in_rp", "land_size_m2", "building_size_m2", "bedrooms", "bathrooms", "carports"]
plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

# JUSTIFICATION

Dari boxplot diatas terlihat ada nilai outliers ekstrem pada harga, luas tanah, dan jumlah kamar. Nilai ini tetap dipertahankan karena masih wajar dalam konteks rumah dengan kategori mewah atau properti komersial sehingga data ini bukan kesalahan. Tetapi analisis sebaiknya menggunakan median atau transformasi log agar tidak bias akibat outlier.

```python
# Low (<1M), Medium (1M-3M), High (3M-10M), Luxury (>10M)
bins = [0, 1000000000, 3000000000, 10000000000, df['price_in_rp'].max()]
labels = ['Low ', 'Medium ', 'High ', 'Luxury ']
df['price_category'] = pd.cut(df['price_in_rp'], bins=bins, labels=labels, include_lowest=True)
print(df[['price_in_rp', 'price_category']].head(10))
print(df['price_category'].value_counts())
```

# B

```python
# 1. Histogram harga rumah
sns.histplot(np.log1p(df['price_in_rp']), bins=40, kde=True)
plt.title("Distribusi Harga Rumah (log)")
plt.show()
```

# INSIGHT

Distribusi harga rumah setelah di-log terlihat lebih seimbang, sebagian besar harga rumah berada di rentang medium (1–3M), sedangkan rumah dengan harga sangat rendah maupun sangat tinggi jumlahnya jauh lebih sedikit. Ini menunjukkan bahwa pasar properti lebih didominasi oleh rumah dengan harga normal atau menengah.

```python
# 2. Barplot kategori harga rumah
plt.figure(figsize=(7,5))
sns.countplot(x='price_category', data=df, order=df['price_category'].value_counts().index)
plt.title("Distribusi Rumah berdasarkan Kategori Harga")
plt.xlabel("Kategori Harga")
plt.ylabel("Jumlah Rumah")
plt.show()
```

# INSIGHT

Kebanyakan dari properti yang dijual adalah properti dengan kategori Medium (1M-3M) dan juga Low (<1M), berbeda jauh jumlahnya dengan High dan Luxury.

```python
# 3. Boxplot harga berdasarkan tipe properti
plt.figure(figsize=(8,5))
sns.boxplot(x='property_type', y=np.log1p(df['price_in_rp']), data=df)
plt.title("Perbandingan Harga Rumah per Tipe Properti (Log Harga)")
plt.xlabel("Tipe Properti")
plt.ylabel("Log Harga Rumah")
plt.show()
```

# INSIGHT

Sebagian besar properti di dataset adalah tipe rumah, dari boxplot terlihat bahwa harga rumah cenderung berkumpul di kisaran menengah sekitar log 20–22, atau sekitar 1–3 miliar rupiah. Ada juga beberapa rumah dengan harga sangat tinggi yang muncul sebagai outlier. Ini menunjukkan bahwa walaupun sebagian besar rumah berada di kisaran harga yang terjangkau sampai menengah, ada beberapa rumah dengan harga jauh lebih mahal, kemungkinan karena lokasi premium atau ukuran tanah yang sangat besar.

```python
# 4. Scatterplot luas tanah vs harga
sns.scatterplot(x=df['land_size_m2'], y=df['price_in_rp'])
plt.yscale("log")
plt.title("Luas Tanah vs Harga Rumah")
plt.show()

```

# INSIGHT

Terlihat ada hubungan positif antara luas tanah dan harga rumah, semakin besar luas tanah, harga rumah juga lebih tinggi. Tetapi persebarannya cukup lebar yang artinya tidak semua rumah dengan tanah luas punya harga sangat mahal. Ada faktor lain seperti lokasi dan fasilitas yang juga memengaruhi harga, beberapa titik yang jauh di atas menunjukkan rumah dengan harga sangat tinggi kemungkinan termasuk kategori luxury.

```python
# 5. Heatmap korelasi antar variabel numerik
plt.figure(figsize=(10,7))
corr = df[['price_in_rp','land_size_m2','building_size_m2','bedrooms','bathrooms','carports']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Korelasi Antar Variabel Numerik")
plt.show()
```

# INSIGHT

Berdasarkan heatmap diatas, harga rumah (price_in_rp) memiliki korelasi sedang dengan luas bangunan (building_size_m2) sebesar 0.49 dan dengan luas tanah (land_size_m2) sebesar 0.37 artinya semakin besar ukuran bangunan atau tanah, semakin tinggi pula harga rumah. Variabel lain seperti jumlah kamar tidur, kamar mandi, maupun carport memiliki korelasi yang lemah terhadap harga.

Selain itu, terlihat hubungan yang sangat kuat antara jumlah kamar tidur dan jumlah kamar mandi dengan nilai korelasi 0.94, yang berarti rumah dengan banyak kamar tidur hampir selalu memiliki banyak kamar mandi. Luas bangunan juga cukup berkorelasi dengan jumlah kamar tidur 0.60 dan kamar mandi 0.64, yang berarti semakin besar bangunan, semakin banyak ruang untuk kamar. Sementara itu, jumlah carport memiliki korelasi rendah dengan semua variabel lain, sehingga tidak terlalu memengaruhi harga maupun ukuran rumah.
